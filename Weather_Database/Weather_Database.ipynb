```python
# Use the citipy module to determine city based on latitude and longitude.
from citipy import citipy

# Import the API key.
from config import weather_api_key
```

```python
# Store the CSV you saved created in part one into a DataFrame.
city_data_df = pd.read_csv("cities.csv")
city_data_df.head()
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
# Create a set of random latitude and longitude combinations.
lats = []
lngs = []

for i in range(2000):
    lats.append(np.random.uniform(low=-90.000, high=90.000))
    lngs.append(np.random.uniform(low=-180.000, high=180.000))

lat_lngs = (lats, lngs)
lat_lngs
```

```python
# Add the latitudes and longitudes to a list.
coordinates = list(lat_lngs[0])
```

```python
len(coordinates)
```

```python
# Import the requests library.
import requests
import json

# Import the API key.
from config import weather_api_key
```

```python
#df = pd.DataFrame(columns=['City','Country','Lat','Lng','Max Temp','Humidity','Cloudiness','Wind Speed','Current Description'])
df = []

for i in range(10):
    city = citipy.nearest_city(lat_lngs[0][i],lat_lngs[1][i])    
    try:
        
        # Starting URL for Weather Map API Call.
        url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key + '&q=' + city.city_name
        #print(url)
        city_weather = requests.get(url)
        #print(json.dumps(city_weather.json(), indent=2))
        
        
        df.append({
            'Lat': city_weather.json()['coord']['lat'],
             
            'Lon': city_weather.json()['coord']['lon'],
            'Max Temp': city_weather.json()['main']['temp_max'],
            'City': city.city_name,
            'Country': city_weather.json()['sys']['country'],
            'Humidity': city_weather.json()['main']['humidity'],
            'Cloudiness': city_weather.json()['clouds']['all'],
            'Wind Speed': city_weather.json()['wind']['speed'],
            'Current Description': city_weather.json()['weather'][0]['description'],
        })
        
        print(df)
    except:
        print("city not found")
        
df = pd.DataFrame(df)
```
